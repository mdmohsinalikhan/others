# Bayesian inference

* Assuming our data follows the model $p(x | \theta)$ with unknown parameters $\theta$, probability theory can be used to derive predictive distribution on new data $x^*$ given observations $X$ as
$$ p(x^* | X) = \int_\Theta p(x^* | \theta) p(\theta | X) \mathrm{d}\theta $$
* This integral depends on $p(\theta | X)$, the *posterior distribution* of the parameters
* Bayes rule:
$$ p(\theta | X) = \frac{p(X | \theta) p(\theta)}{p(X)} = \frac{p(X | \theta) p(\theta)}{\int_{\boldsymbol{\Theta}} p(X | \theta) p(\theta) \mathrm{d}\theta} $$
* Given samples from $p(\theta | X)$ we could evaluate $p(x^* | X)$ using Monte Carlo integration
* Intractable denominator $p(X)$ - sampling from $$ P(x) = \frac{P^*(x)}{Z}, $$
where $Z$ is unknown while $P^*(x)$ is known

# Markov chain Monte Carlo

* Instead of direct sampling, we develop a Markov chain that converges to the correct distribution
* Markov chain: $ p(x_t | x_1, \dots, x_{t-1}) = p(x_t | x_{t-1}) $
* Defined for a sequence $x_1, \dots, x_n$ defined by the initial distribution $p(x_1)$ and transition kernel $p(x_t | x_{t-1})$

# Markov chain theory

* Irreducibility - can reach any state from any other with positive probability in a finite number of steps
* Periodicity / **aperiodicity**
* Ergodicity: Aperiodic and irreducible with finite expected return time to every state
* Stationarity: $p(x_{t+1}, \dots, x_{t+k})$ does not depend on $t$
* Reversibility: $(x_t, x_{t+1})$ exchangeable for all $t$ - implies stationarity
* Sufficient condition for reversibility: detailed balance
  $$ \int_A \pi(\theta) K(\theta, B) \mathrm{d}\theta = \int_B \pi(\phi) K(\phi, A) \mathrm{d}\phi \quad \forall A, B \subset S. $$
* A reversible ergodic chain has a unique stationary distribution

# Metropolis-Hastings algorithm

MH sampling works by deriving a Markov chain $x^{(1)}, x^{(2)}, \dots$ whose stationary distribution is the target $P(x)$. The algorithm works by proposing a new state $x'$ from a proposal density $Q(x' ; x^{(t)})$. The proposal is either accepted or rejected based on the quantity
$$ a = \frac{P^*(x')}{P^*(x^{(t)})} \frac{Q(x^{(t)} ; x')}{Q(x' ; x^{(t)})} $$
as follows:
* If $a \ge 1$ then the new state is accepted.
* Otherwise, the new state is accepted with probability $a$.

If the step is accepted, we set $x^{(t+1)} = x'$.  
If the step is rejected, then we set $x^{(t+1)} = x^{(t)}$.  

The MH update fulfills detailed balance and the chain is thus reversible. Ergodic with a suitable proposal.

```python
%matplotlib inline
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

# Metropolis sampling (symmetric proposal) for given log-target distribution
def mhsample(x0, n, logtarget, drawproposal):
    x = x0
    xs = np.zeros(n)
    for i in range(n):
        x_prop = drawproposal(x)
        if npr.rand() < np.exp(logtarget(x_prop) - logtarget(x)):
            x = x_prop
        xs[i] = x
    return xs

# Testing Laplace target with normal proposal
x = mhsample(0.0, 10000, lambda x: -np.abs(x), lambda x: x+npr.normal())
h = plt.hist(x[0:10000:10], 50, normed=True)
plt.show()
plt.plot(x[0:10000:10])
```

```python
# Example of a poorly converging MCMC run with too short jumps
x = mhsample(0.0, 10000, lambda x: -np.abs(x), lambda x: x+0.1*npr.normal())
h = plt.hist(x[0:10000:10], 50, normed=True)
plt.show()
plt.plot(x[0:10000:10])
```

```python
# Example of a poorly converging MCMC run with too long jumps
x = mhsample(0.0, 10000, lambda x: -np.abs(x), lambda x: x+50*npr.normal())
h = plt.hist(x[0:10000:10], 50, normed=True)
plt.show()
plt.plot(x[0:10000:10])
```
